```python
genecounts = pd.read_csv('SkinCancer_genecounts.txt', sep='\t', index_col=0)
```

```python
def order_data(raw_data):
    '''order patients in spesific order for consistency.'''
    bcc = {'12-2':'Intensity 12_C1_lysis', '12-3':'Intensity 12_C3_lysis', '12-5':'Intensity 12_C5_lysis',
           '27-2':'Intensity 27_C1_lysis', '27-3':'Intensity 27_C3_lysis', '27-5':'Intensity 27_C5_lysis', 
           '39-1':'Intensity 39_C1_lysis','39-3':'Intensity 39_C3_lysis', '39-5':'Intensity 39_C5_lysis'}

    scc = {'21-2':'Intensity 21_C1_lysis', '21-4':'Intensity 21_C3_lysis', '21-5':'Intensity 21_C5_lysis',
           '31-1':'Intensity 31_C1_lysis', '31-3':'Intensity 31_C3_lysis', '31-5':'Intensity 31_C5_lysis', 
           '32-1':'Intensity 32_C1_lysis', '32-3':'Intensity 32_C3_lysis', '32-5':'Intensity 32_C5_lysis'}

    bcc_scc = {'25-1':'Intensity 25_C1_lysis', '25-3':'Intensity 25_C3_lysis', '25-5':'Intensity 25_C5_lysis'}

    keratosis = {'33-1':'Intensity 33_C1_lysis', '33-3':'Intensity 33_C3_lysis', '33-5':'Intensity 33_C5_lysis'}
    
    patients = [key for key,val in bcc.items()]
    print(patients)
    patients.extend([key for key,val in scc.items()])
    print(patients)
    patients.extend([key for key,val in bcc_scc.items()])
    patients.extend([key for key,val in keratosis.items()])
#     print(patients)
    df = raw_data.copy()
    df = df.reindex(columns=patients)
    df.columns = ['{0}'.format(f'BCC {bcc[c]}' if c in bcc else f'SCC {scc[c]}' if c in scc else f'SCC&BCC {bcc_scc[c]}' if c in bcc_scc else f'keratosis {keratosis[c]}') for c in df.columns]
    return df 
```

```python
ordered_genecounts = order_data(genecounts)
ordered_genecounts
```
